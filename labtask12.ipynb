```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd
train_data = pd.read_csv("/content/drive/MyDrive/train.csv")
train_data.head()
test_data = pd.read_csv("/content/drive/MyDrive/test.csv")
test_data.head()
```

```python
full_data = pd.concat([train_data, test_data], sort=False)
full_data.head()
```

**Read Csv file**

```python
full_data['Age'].fillna(full_data['Age'].median(), inplace = True)
full_data['Fare'].fillna(full_data['Fare'].median(), inplace = True)
```

```python
label = LabelEncoder()
full_data['Sex'] = label.fit_transform(full_data['Sex'])
full_data['Embarked'] = label.fit_transform(full_data['Embarked'])
```

```python
columns_to_drop = ['Name', 'Ticket', 'Cabin']
existing_columns = [col for col in columns_to_drop if col in full_data.columns]
full_data.drop(existing_columns, axis=1, inplace=True)
train_data = full_data[full_data['Survived'].notna()]
test_data = full_data[full_data['Survived'].isna()]
x_train = train_data.drop(['Survived', 'PassengerId'], axis=1)
y_train = train_data['Survived']
x_test = test_data.drop(['Survived', 'PassengerId'], axis=1)

x_train_split, X_val, y_train_split, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
x_train_split = scaler.fit_transform(x_train_split)
x_val = scaler.transform(X_val)
x_test = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train_split, y_train_split)
```

```python
y_pred = rf_model.predict(X_val)
accuracy_score(y_val, y_pred)
```

```python
 print(classification_report(y_val, y_pred))
```
